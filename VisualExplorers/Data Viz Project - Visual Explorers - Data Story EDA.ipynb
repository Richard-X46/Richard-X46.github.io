```python
import pandas as pd
import plotly.express as px

file = "Datasets/AVG_salaries_based_on_prof.csv"
df = pd.read_csv(file)

# Convert REF_DATE to datetime
df["REF_DATE"] = pd.to_datetime(df["REF_DATE"], format="%Y-%m")
df["REF_DATE_STR"] = df["REF_DATE"].dt.strftime("%Y-%m")

# Sort by date for proper animation
df = df.sort_values("REF_DATE")

# Create cumulative data for each frame
frames_data = []
for date in df["REF_DATE_STR"].unique():
    frame_df = df[df["REF_DATE_STR"] <= date].copy()
    frame_df["frame"] = date
    frames_data.append(frame_df)

df_animated = pd.concat(frames_data, ignore_index=True)

# Animated line plot with cumulative data
fig = px.line(
    df_animated,
    x="REF_DATE",
    y="VALUE",
    color="North American Industry Classification System (NAICS)",
    animation_frame="frame",
    markers=True,
    title="Animated Average Hourly Wages by Industry Over Time",
    labels={
        "VALUE": "Average Hourly Wage (CAD)",
        "REF_DATE": "Date",
        "North American Industry Classification System (NAICS)": "Industry"
    },
    range_y=[df["VALUE"].min() * 0.95, df["VALUE"].max() * 1.05]  # Fixed y-axis range
)

fig.update_layout(
    height=700, 
    width=1200, 
    legend_title_text="Industry",
    xaxis_range=[df["REF_DATE"].min(), df["REF_DATE"].max()]  # Fixed x-axis range
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

file = "Datasets/AVG_salaries_based_on_prof.csv"
df = pd.read_csv(file)

# Convert REF_DATE to datetime
df["REF_DATE"] = pd.to_datetime(df["REF_DATE"], format="%Y-%m")
df["REF_DATE_STR"] = df["REF_DATE"].dt.strftime("%Y-%m")

# filter for specific industries if needed
industries_of_interest = ["Transportation and warehousing [48-49]","Professional, scientific and technical services [54,541]"]
df = df[df["North American Industry Classification System (NAICS)"].isin(industries_of_interest)]

# Sort by date for proper animation
df = df.sort_values("REF_DATE")

# Create cumulative data for each frame
frames_data = []
for date in df["REF_DATE_STR"].unique():
    frame_df = df[df["REF_DATE_STR"] <= date].copy()
    frame_df["frame"] = date
    frames_data.append(frame_df)

df_animated = pd.concat(frames_data, ignore_index=True)

# Animated line plot with cumulative data
fig = px.line(
    df_animated,
    x="REF_DATE",
    y="VALUE",
    color="North American Industry Classification System (NAICS)",
    animation_frame="frame",
    markers=True,
    title="Animated Average Hourly Wages by Industry Over Time",
    labels={
        "VALUE": "Average Hourly Wage (CAD)",
        "REF_DATE": "Date",
        "North American Industry Classification System (NAICS)": "Industry"
    },
    range_y=[df["VALUE"].min() * 0.95, df["VALUE"].max() * 1.05]
)

fig.update_layout(
    height=700, 
    width=1200, 
    legend_title_text="Industry",
    xaxis_range=[df["REF_DATE"].min(), df["REF_DATE"].max()],
    # Control animation speed here
    updatemenus=[{
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 100, "redraw": True},
                                "fromcurrent": True, 
                                "transition": {"duration": 50}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": True},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }]
)

# Set default animation speed
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 100
fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 50

fig.show()
```

# AI Exposure Quadrant Chart for Jiya vs Chris
Plotting AI Occupational Exposure (AIOE) vs Complementarity to show which jobs are enhanced vs replaced by AI

```python
import plotly.graph_objects as go
import numpy as np

# Jiya and Chris data points
personas = {
    'Jiya (AI Engineer)': {
        'category': 'Computer and information systems professionals',
        'aioe': 6.5877,
        'complementarity': 0.5513,
        'color': '#1f77b4'
    },
    'Chris (Logistics Supervisor)': {
        'category': 'Sales and service supervisors',
        'aioe': 6.0866,
        'complementarity': 0.604,
        'color': '#ff7f0e'
    }
}

# Calculate median values for quadrant lines (from the full dataset)
median_aioe = 6.0758  # Total average from data
median_comp = 0.5953  # Total average from data

# Create figure with flipped axes (AIOE on x, complementarity on y)
fig_aioe = go.Figure()

# Add quadrant background rectangles (flipped)
fig_aioe.add_shape(type="rect", x0=5.0, y0=median_comp, x1=median_aioe, y1=0.75,
              fillcolor="rgba(200,255,200,0.2)", line_width=0, layer="below")  # Low exposure, high comp
fig_aioe.add_shape(type="rect", x0=median_aioe, y0=median_comp, x1=7.0, y1=0.75,
              fillcolor="rgba(255,255,200,0.2)", line_width=0, layer="below")  # High exposure, high comp
fig_aioe.add_shape(type="rect", x0=5.0, y0=0.5, x1=median_aioe, y1=median_comp,
              fillcolor="rgba(230,230,230,0.2)", line_width=0, layer="below")  # Low exposure, low comp
fig_aioe.add_shape(type="rect", x0=median_aioe, y0=0.5, x1=7.0, y1=median_comp,
              fillcolor="rgba(255,200,200,0.2)", line_width=0, layer="below")  # High exposure, low comp

# Add quadrant dividing lines
fig_aioe.add_vline(x=median_aioe, line_dash="dash", line_color="gray", opacity=0.5)
fig_aioe.add_hline(y=median_comp, line_dash="dash", line_color="gray", opacity=0.5)

# Add persona data points
for name, data in personas.items():
    fig_aioe.add_trace(go.Scatter(
        x=[data['aioe']],
        y=[data['complementarity']],
        mode='markers+text',
        name=name,
        marker=dict(size=20, color=data['color'], line=dict(width=2, color='white')),
        text=[name.split('(')[0].strip()],
        textposition="top center",
        textfont=dict(size=12, color=data['color'], family='Inter'),
        hovertemplate=f"<b>{name}</b><br>" +
                      f"Category: {data['category']}<br>" +
                      f"AI Exposure: {data['aioe']:.2f}<br>" +
                      f"Complementarity: {data['complementarity']:.2f}<br>" +
                      "<extra></extra>"
    ))

# Add quadrant labels (adjusted for flipped axes)
fig_aioe.add_annotation(x=5.5, y=0.68, text="Low Risk<br>(Stable)", 
                   showarrow=False, font=dict(size=9, color="green"), opacity=0.6)
fig_aioe.add_annotation(x=6.6, y=0.68, text="Opportunity<br>(AI Enhances)", 
                   showarrow=False, font=dict(size=9, color="#228B22"), opacity=0.6)
fig_aioe.add_annotation(x=5.5, y=0.54, text="Less Affected", 
                   showarrow=False, font=dict(size=9, color="gray"), opacity=0.6)
fig_aioe.add_annotation(x=6.6, y=0.54, text="High Risk<br>(AI Replaces)", 
                   showarrow=False, font=dict(size=9, color="red"), opacity=0.6)

# Update layout
fig_aioe.update_layout(
    title="AI Impact: Complementarity vs Exposure",
    xaxis_title="AI Occupational Exposure (AIOE)",
    yaxis_title="Complementarity Index",
    xaxis=dict(range=[5.0, 7.0], showgrid=True, gridcolor='lightgray'),
    yaxis=dict(range=[0.5, 0.75], showgrid=True, gridcolor='lightgray'),
    height=400,
    width=600,
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Inter, sans-serif', color='#262730'),
    showlegend=False
)

fig_aioe.show()
```

# Interactive Dynamic Quadrant Chart
Create a fully interactive quadrant chart with filters and multi-select capabilities

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Load the AI exposure data
df_ai = pd.read_csv("Datasets/AI_Exposure_data.csv")

# Clean column names
df_ai.columns = df_ai.columns.str.strip()

# Display first few rows to verify
print("Columns:", df_ai.columns.tolist())
print("\nFirst 5 rows:")
df_ai.head()
```

```python
# Create interactive quadrant chart with dropdown filters
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Remove the "Total" row and clean data
df_filtered = df_ai[df_ai['Job Category'] != 'Total'].copy()

# Convert percentage columns to numeric (remove % sign and convert)
percentage_cols = ['High exposure, low complementarity', 'High exposure, high complementarity', 'Low exposure']
for col in percentage_cols:
    df_filtered[col] = pd.to_numeric(df_filtered[col].astype(str).str.replace('%', ''), errors='coerce')

# Calculate median values for quadrant lines
median_aioe = df_ai[df_ai['Job Category'] == 'Total']['AIOE'].values[0]
median_comp = df_ai[df_ai['Job Category'] == 'Total']['Potential complementarity'].values[0]

# Create color mapping based on exposure categories
def get_color(row):
    high_exp_low_comp = row['High exposure, low complementarity']
    high_exp_high_comp = row['High exposure, high complementarity']
    low_exp = row['Low exposure']
    
    if high_exp_low_comp > 50:
        return '#ff6b6b'  # Red - High risk
    elif high_exp_high_comp > 50:
        return '#51cf66'  # Green - Opportunity
    elif low_exp > 50:
        return '#339af0'  # Blue - Low exposure
    else:
        return '#ffd43b'  # Yellow - Mixed

df_filtered['color'] = df_filtered.apply(get_color, axis=1)

# Create figure with dropdown menus
fig = go.Figure()

# Add all points initially
for idx, row in df_filtered.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['AIOE']],
        y=[row['Potential complementarity']],
        mode='markers+text',
        name=row['Job Category'],
        marker=dict(
            size=np.sqrt(row['Employment']) / 150,  # Size based on employment
            color=row['color'],
            line=dict(width=2, color='white'),
            opacity=0.8
        ),
        text=[row['Job Category'].split('(')[0][:20]],  # Shortened labels
        textposition="top center",
        textfont=dict(size=9, family='Inter'),
        hovertemplate=(
            f"<b>{row['Job Category']}</b><br>" +
            f"AIOE: {row['AIOE']:.2f}<br>" +
            f"Complementarity: {row['Potential complementarity']:.2f}<br>" +
            f"Employment: {row['Employment']:,.0f}<br>" +
            f"High Risk %: {row['High exposure, low complementarity']:.0f}%<br>" +
            f"Opportunity %: {row['High exposure, high complementarity']:.0f}%<br>" +
            "<extra></extra>"
        ),
        visible=True
    ))

# Add quadrant background shapes
fig.add_shape(type="rect", x0=5.0, y0=median_comp, x1=median_aioe, y1=0.75,
              fillcolor="rgba(81,207,102,0.15)", line_width=0, layer="below")
fig.add_shape(type="rect", x0=median_aioe, y0=median_comp, x1=7.0, y1=0.75,
              fillcolor="rgba(255,212,59,0.15)", line_width=0, layer="below")
fig.add_shape(type="rect", x0=5.0, y0=0.45, x1=median_aioe, y1=median_comp,
              fillcolor="rgba(51,154,240,0.15)", line_width=0, layer="below")
fig.add_shape(type="rect", x0=median_aioe, y0=0.45, x1=7.0, y1=median_comp,
              fillcolor="rgba(255,107,107,0.15)", line_width=0, layer="below")

# Add quadrant dividing lines
fig.add_vline(x=median_aioe, line_dash="dash", line_color="gray", opacity=0.5, line_width=2)
fig.add_hline(y=median_comp, line_dash="dash", line_color="gray", opacity=0.5, line_width=2)

# Add quadrant labels
fig.add_annotation(x=5.4, y=0.68, text="<b>Stable & Safe</b><br>Low Risk", 
                   showarrow=False, font=dict(size=11, color="#51cf66"), opacity=0.7)
fig.add_annotation(x=6.7, y=0.68, text="<b>Opportunity</b><br>AI Enhances", 
                   showarrow=False, font=dict(size=11, color="#ffd43b"), opacity=0.7)
fig.add_annotation(x=5.4, y=0.52, text="<b>Less Affected</b><br>Low Exposure", 
                   showarrow=False, font=dict(size=11, color="#339af0"), opacity=0.7)
fig.add_annotation(x=6.7, y=0.52, text="<b>High Risk</b><br>AI Replaces", 
                   showarrow=False, font=dict(size=11, color="#ff6b6b"), opacity=0.7)

# Create dropdown filter buttons for risk categories
buttons = [
    dict(label="All Occupations",
         method="update",
         args=[{"visible": [True] * len(df_filtered)},
               {"title": "AI Impact Across All Occupations"}]),
    dict(label="High Risk Only",
         method="update",
         args=[{"visible": [row['High exposure, low complementarity'] > 50 for _, row in df_filtered.iterrows()]},
               {"title": "High Risk Occupations (AI May Replace)"}]),
    dict(label="High Opportunity Only",
         method="update",
         args=[{"visible": [row['High exposure, high complementarity'] > 50 for _, row in df_filtered.iterrows()]},
               {"title": "High Opportunity Occupations (AI Enhances)"}]),
    dict(label="Low Exposure Only",
         method="update",
         args=[{"visible": [row['Low exposure'] > 50 for _, row in df_filtered.iterrows()]},
               {"title": "Low Exposure Occupations"}]),
    dict(label="Management Only",
         method="update",
         args=[{"visible": ['Management' in row['Job Category'] or 'management' in row['Job Category'].lower() 
                           for _, row in df_filtered.iterrows()]},
               {"title": "Management Occupations"}]),
    dict(label="Tech/Professional",
         method="update",
         args=[{"visible": [any(word in row['Job Category'].lower() for word in ['computer', 'professional', 'technical', 'engineer', 'science']) 
                           for _, row in df_filtered.iterrows()]},
               {"title": "Tech & Professional Occupations"}]),
]

# Update layout with interactive controls
fig.update_layout(
    title="Interactive AI Impact Quadrant: All Canadian Occupations",
    xaxis_title="AI Occupational Exposure (AIOE)",
    yaxis_title="Complementarity Index (Higher = AI Enhances Job)",
    xaxis=dict(range=[5.0, 7.0], showgrid=True, gridcolor='#e0e0e0'),
    yaxis=dict(range=[0.45, 0.75], showgrid=True, gridcolor='#e0e0e0'),
    height=700,
    width=1000,
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Inter, sans-serif', color='#262730'),
    showlegend=False,
    updatemenus=[
        dict(
            buttons=buttons,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.02,
            xanchor="left",
            y=1.15,
            yanchor="top",
            bgcolor='white',
            bordercolor='#ccc',
            borderwidth=1
        )
    ],
    annotations=[
        dict(
            text="<b>Filter by Category:</b>",
            showarrow=False,
            x=0.02, xanchor="left", xref="paper",
            y=1.19, yanchor="bottom", yref="paper",
            font=dict(size=12)
        )
    ]
)

fig.show()

# Print summary statistics
print("\n Summary Statistics:")
print(f"Total Occupations: {len(df_filtered)}")
print(f"High Risk (>50%): {len(df_filtered[df_filtered['High exposure, low complementarity'] > 50])}")
print(f"High Opportunity (>50%): {len(df_filtered[df_filtered['High exposure, high complementarity'] > 50])}")
print(f"Low Exposure (>50%): {len(df_filtered[df_filtered['Low exposure'] > 50])}")
```
